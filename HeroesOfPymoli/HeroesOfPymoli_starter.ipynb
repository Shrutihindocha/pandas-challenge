```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
```

## Player Count

* Display the total number of players


```python
player_count = pd.Series(len(purchase_data["SN"].value_counts()))
player_count

player_count_df = pd.DataFrame({"Total Players" : player_count})

player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
no_of_unique_items = len(purchase_data["Item Name"].value_counts())
# no_of_unique_items 
avg_price = f'${round(purchase_data["Price"].mean(),2)}'
# avg_price 
no_of_purchases = purchase_data["Purchase ID"].count()
# no_of_purchases
total_revenue = f'${purchase_data["Price"].sum()}'
# total_revenue

purchasing_analysis = pd.DataFrame([{"Number of Unique Items" : no_of_unique_items,
                        "Average Price" : avg_price,
                        "Number of Purchases" : no_of_purchases,
                        "Total Revenue" : total_revenue}])

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_demographics = purchase_data.loc[:, ["Gender", "SN", "Age"]]
player_demographics = player_demographics.drop_duplicates()

gender_demographics = pd.Series(player_demographics["Gender"].value_counts())

gender_demographics_df = pd.DataFrame(gender_demographics)

gender_demographics_df
```

```python
#Calculating total number of players for all gender
sum_of_all_players = gender_demographics_df["Gender"].sum()

sum_of_all_players

#adding column for % of players
gender_demographics_df["Percentage of Players"]= round(gender_demographics_df["Gender"]/sum_of_all_players*100,2)

gender_demographics_df
```

```python
#renaming column name to "Total Count"
gender_demographics_df = gender_demographics_df.rename(columns = {"Gender" : "Total Count"})

gender_demographics_df
```

```python
#Adding % symbol

gender_demographics_df["Percentage of Players"]= gender_demographics_df["Percentage of Players"].astype(str) + '%'
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase count

#dropping duplicates
purchase_data_gender_count = purchase_data.drop_duplicates()

#Grouping
purchase_data_gender_count = purchase_data.groupby(["Gender"])

#Filtering for gender count using Purchase ID
purchase_data_gender_count = pd.Series(purchase_data_gender_count.count()["Purchase ID"])

purchase_data_gender_count
```

```python
#Avg Purchase Price
avg_purchase_price = purchase_data.groupby(["Gender"]).mean()

#Filtering for Price
avg_purchase_price = '$' + round(avg_purchase_price["Price"],2).astype(str)

avg_purchase_price = pd.Series(avg_purchase_price)

avg_purchase_price
```

```python
#Total purchase value

total_purchase_value = purchase_data.groupby(["Gender"]).sum()

#Filtering for Price
total_purchase_value = '$' + (round(total_purchase_value["Price"],2)).astype(str)

total_purchase_value = pd.Series(total_purchase_value)

total_purchase_value
```

```python
#Groupby SN
total_purchase_pp = purchase_data.groupby(['Gender','SN']).sum()

total_purchase_pp["Price"]

# Grouping the above again by Gender

avg_total_purchase_pp = '$'+(round(total_purchase_pp["Price"].groupby('Gender').mean(),2)).astype(str)

avg_total_purchase_pp = pd.Series(avg_total_purchase_pp)

avg_total_purchase_pp
```

```python
#Creating new DataFrame
Purchasing_analysis_gender = pd.DataFrame({'Purchase Count' : purchase_data_gender_count,
                                           'Average Purchase Price' : avg_purchase_price,
                                           'Total Purchase Price' : total_purchase_value,
                                           'Avg Total Purchase per Person' : avg_total_purchase_pp})

Purchasing_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establishing bins for age

#Checking the range
min = purchase_data["Age"].min()
max = purchase_data["Age"].max()

# print(min)
# print(max)

#Creating bins
bins = [0,9,14,19,24,29,34,39,10000]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

age_group = pd.cut(purchase_data["Age"], bins, labels = labels)

purchase_data["Age Groups"] = age_group

purchase_data
```

```python
# Calculating the numbers and percentages by age group

purchase_data_AG = purchase_data.drop_duplicates(["Gender", "SN", "Age"])
purchase_data_AG

# Create a summary data frame to hold the results
purchase_data_age_group = purchase_data_AG.groupby(["Age Groups"])

total_count = purchase_data_age_group["Purchase ID"].count()

total_count
```

```python
# Rounding the percentage column to two decimal places and formating the output
percent_of_players = round((total_count/(total_count.sum()))*100,2).astype(str) + '%'

percent_of_players
```

```python
# Create a summary data frame to hold the results
age_demo = pd.DataFrame({'Total Count':total_count,
                        'Percentage of Players':percent_of_players})

# Displaying Age Demographics Table
age_demo 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Perform calculations to obtain purchase count, item price, and total purchase value
purchase_count_MPI = purchase_data.groupby(["Item ID", "Item Name"])["Price"]

purchase_count_MPI = pd.Series(purchase_count_MPI.count())
purchase_count_MPI = purchase_count_MPI.sort_index()

purchase_count_MPI.sort_values(ascending=False)
```

```python
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. 

most_popular_items = purchase_data.drop_duplicates(subset = ['Item Name'])
most_popular_items = most_popular_items.set_index(["Item ID", "Item Name"])
most_popular_items

item_price = most_popular_items["Price"].sort_index()

item_price = item_price

item_price
```

```python
total_purchase_value = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].sort_index()

total_purchase_value = pd.Series(total_purchase_value)

total_purchase_value
```

```python
# Create a summary data frame to hold the results
most_popular = pd.DataFrame({'Purchase Count' : purchase_count_MPI,
                                           'Item Price' : item_price,
                                           'Total Purchase Value' : total_purchase_value})

most_popular
```

```python
# Sort the purchase count column in descending order
most_popular.sort_values("Purchase Count", ascending=False).head()
```
